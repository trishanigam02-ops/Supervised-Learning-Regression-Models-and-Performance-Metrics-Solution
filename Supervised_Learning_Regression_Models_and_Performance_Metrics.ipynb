# Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.
- Simple Linear Regression (SLR) is a statistical technique used to study the relationship between one independent variable (X) and one dependent variable (Y). It helps in understanding how changes in the independent variable influence the dependent variable.

The main purpose of Simple Linear Regression is to model, explain, and predict the value of the dependent variable based on the independent variable. It assumes that the relationship between X and Y can be represented by a straight line.

Mathematical form:
Y=a+bX

Where:

Y is the dependent variable

X is the independent variable

a is the intercept (value of Y when X = 0)

b is the slope (change in Y for one unit change in X)

Purpose of Simple Linear Regression:

To identify the strength and direction of the relationship between two variables.

To predict future values of the dependent variable.

To analyze trends and patterns in data.

In real-world applications, SLR is widely used in fields like business, economics, marketing, and data analytics for forecasting and decision-making.

# Question 2: What are the key assumptions of Simple Linear Regression?
- The key assumptions of Simple Linear Regression are the conditions that must be satisfied to ensure the model produces reliable and accurate results. These assumptions are explained below:

Linearity
There must be a linear relationship between the independent variable (X) and the dependent variable (Y). This means that changes in X lead to proportional changes in Y.

Independence of Errors
The observations and error terms should be independent of each other. One observation should not influence another.

Homoscedasticity
The variance of the error terms should be constant across all values of X. In other words, the residuals should have equal spread around the regression line.

Normality of Errors
The residuals (errors) should be normally distributed. This assumption is important for conducting hypothesis tests and building confidence intervals.

No Perfect Multicollinearity
Since simple linear regression has only one independent variable, this assumption is automatically satisfied. It means the independent variable should not be perfectly correlated with other predictors.

When these assumptions hold true, the Simple Linear Regression model provides valid interpretations and accurate predictions.

# Question 3: Write the mathematical equation for a simple linear regression model and explain each term.
- The mathematical equation for a Simple Linear Regression model is:
Y=a+bX

Where each term has the following meaning:

Y: The dependent variable, which is the outcome or response we want to predict (for example, sales, marks, or price).

X: The independent variable, which is the predictor or input variable that influences Y (for example, hours studied or advertising cost).

a (Intercept): The value of Y when X is equal to 0. It represents the starting point of the regression line on the Y-axis.

b (Slope): The rate at which Y changes with respect to X. It shows how much Y increases or decreases when X increases by one unit.

In simple terms, this equation represents a straight-line relationship between X and Y, and it is used to explain patterns in data and make predictions.

# Question 6: What is Logistic Regression? How does it differ from Linear Regression?
- | Feature          | Linear Regression    | Logistic Regression      |
| ---------------- | -------------------- | ------------------------ |
| Type of output   | Continuous values    | Categorical (binary)     |
| Example output   | Price, marks, salary | Yes/No, 0/1              |
| Model line       | Straight line        | S-shaped (sigmoid curve) |
| Prediction range | Any real number      | Between 0 and 1          |
| Use case         | Regression problems  | Classification problems  |


Linear Regression predicts numbers

Logistic Regression predicts probabilities and classes

Logistic regression is widely used in classification tasks such as fraud detection, medical diagnosis and customer behavior prediction.

# Question 7: Name and briefly describe three common evaluation metrics for regression models.

- Three common evaluation metrics used to assess the performance of regression models are explained below:

Mean Absolute Error (MAE)
MAE measures the average absolute difference between the actual values and the predicted values. It shows how far predictions are from actual outcomes on average. Lower MAE indicates better model performance.

Mean Squared Error (MSE)
MSE calculates the average of the squared differences between actual and predicted values. It penalizes larger errors more heavily than smaller ones, making it useful when large errors are particularly undesirable.

Root Mean Squared Error (RMSE)
RMSE is the square root of MSE and represents the error in the same units as the dependent variable. It is easy to interpret and widely used to evaluate regression models.

These metrics help data scientists understand prediction accuracy and compare different regression models effectively.


# Question 8: What is the purpose of the R-squared metric in regression analysis?

- The purpose of the R-squared (RÂ²) metric in regression analysis is to measure how well the independent variable(s) explain the variation in the dependent variable.

In simple words, RÂ² tells us how good the model fits the data.

Key Points:

RÂ² represents the proportion of variance in the dependent variable that is explained by the model.

Its value ranges from 0 to 1.

RÂ² = 0 â†’ Model explains none of the variation.

RÂ² = 1 â†’ Model explains all the variation.

A higher RÂ² value indicates a better fit of the regression model.

Example:

If RÂ² = 0.85, it means 85% of the variability in the target variable is explained by the independent variables, while the remaining 15% is due to other factors or noise.

Important Note:

A high RÂ² does not always mean the model is perfect.

It does not detect overfitting, multicollinearity, or heteroscedasticity.

RÂ² should be evaluated along with other metrics like MAE, MSE, RMSE, and residual plots.

This makes R-squared a useful but not standalone metric for regression evaluation.

# Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.(Include your Python code and output in the code box below.)


```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
